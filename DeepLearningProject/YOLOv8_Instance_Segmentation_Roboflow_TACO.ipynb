```python
# yaml 파일을 학습이 가능하도록 경로 설정.
# key-value 데이터인 dict 데이터타입으로 data['train'], data['val'], data['nc'], data['names'] 에 넣어주는데,
# 가장 중요한 부분은 데이터 경로 설정임.

import yaml

data = { 'train' : '/content/TACO/train/images/',
         'val' : '/content/TACO/valid/images/',
         'test' : '/content/TACO/test/images',
         'names' : ['trash'],
         'nc' : 1 }

with open('/content/TACO/TACO.yaml', 'w') as f:
  yaml.dump(data, f)


with open('/content/TACO/TACO.yaml', 'r') as f:
  taco_yaml = yaml.safe_load(f)
  display(taco_yaml)
```

# Train (Python version)

```python
# Load YOLOv8n-seg

from ultralytics import YOLO

model = YOLO('yolov8n-seg.pt')  # load a pretrained YOLOv8n segmentation model
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
model.train(data='/content/TACO/TACO.yaml', epochs=50, patience=10, batch=32, imgsz=416)
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

# 테스트 이미지 데이터 생성 및 확인

```python
# 테스트 이미지

from glob import glob

test_image_list = glob('/content/TACO/test/images/*')

print(len(test_image_list))

test_image_list.sort()


for i in range(len(test_image_list)):

    print('i = ',i, test_image_list[i])
```

# 이미지내의 Instance Segmentation (Inference)

```python
results = model.predict(source='/content/TACO/test/images/', save=True)
```

```python
print(type(results), len(results))
```

```python
import numpy as np

for result in results:

    uniq, cnt = np.unique(result.boxes.cls.cpu().numpy(), return_counts=True)  # Torch.Tensor -> numpy
    uniq_cnt_dict = dict(zip(uniq, cnt))

    print('\n{class num:counts} =', uniq_cnt_dict,'\n')

    for c in result.boxes.cls:
        print('class num =', int(c), ', class_name =', model.names[int(c)])
```

# 결과 확인 및 다운로드

```python
import glob

detetced_image_list = glob.glob(('/content/runs/segment/predict/*'))

detected_image_nums = len(detetced_image_list)

print(detected_image_nums)

print(detetced_image_list)
```

```python
# 다운로드를 위한 inference image 압축

import zipfile
import os

if not os.path.exists('/content/detected_result/'):
    os.mkdir('/content/detected_result/')
    print('detected_result dir is created !!!')


with zipfile.ZipFile('/content/detected_result/detected_images.zip', 'w') as detected_images:

    for idx in range(detected_image_nums):
        detected_images.write(detetced_image_list[idx])
```
